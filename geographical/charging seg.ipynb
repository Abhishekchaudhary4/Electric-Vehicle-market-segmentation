```python
def get_data(df):
    """
    This function helps to get the required data
    
    """
    d_mh = pd.read_excel(df,skiprows = [0,1]) # skipping first 2 rows
    d_mh.rename(columns=d_mh.iloc[0], inplace = True) # renaming 1st column
    d_mh = d_mh.iloc[1:,1:]
    d_mh.rename(columns = {' Vehicle Class ': "Vehicle Class"}, inplace = True)
    d_mh = d_mh[["Vehicle Class","CNG ONLY","DIESEL","ELECTRIC(BOV)","PETROL"]] # selecting main fuel types
    return d_mh
```

```python
columns_to_convert = df.columns[1:]
df[columns_to_convert] = df[columns_to_convert].apply(lambda x: x.str.replace(',', '', regex=True).astype(float))
```

```python
df.info()
```

```python
electric_vehicles = df[["Vehicle Class", "ELECTRIC(BOV)"]]

# Convert the "ELECTRIC(BOV)" column to float

# Find the vehicle class with the maximum count of electric vehicles
max_electric_vehicle_class = electric_vehicles[electric_vehicles["ELECTRIC(BOV)"] == electric_vehicles["ELECTRIC(BOV)"].max()]

# Display the vehicle class with the maximum count of electric vehicles
print("Vehicle class with the maximum count of electric vehicles:")
print(max_electric_vehicle_class)
```

```python
plt.figure(figsize=(12, 12))
plt.bar(electric_vehicles["Vehicle Class"], electric_vehicles["ELECTRIC(BOV)"])
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

# Highlight the vehicle class with the maximum count
max_class = max_electric_vehicle_class["Vehicle Class"].values[0]
max_count = max_electric_vehicle_class["ELECTRIC(BOV)"].values[0]
plt.annotate(f'Max: {max_count}', (max_class, max_count), textcoords="offset points", xytext=(0, 10), ha='center')

# Label the axes and title
plt.xlabel("Vehicle Class")
plt.ylabel("Count of Electric Vehicles")
plt.title("Count of Electric Vehicles by Vehicle Class")

# Show the plot
plt.tight_layout()
plt.show()
```

```python
mean_electric_vehicles = df['ELECTRIC(BOV)'].mean()
median_electric_vehicles = df['ELECTRIC(BOV)'].median()
std_dev_electric_vehicles = df['ELECTRIC(BOV)'].std()

# Display the summary statistics
print("Mean Electric Vehicles:", mean_electric_vehicles)
print("Median Electric Vehicles:", median_electric_vehicles)
print("Standard Deviation of Electric Vehicles:", std_dev_electric_vehicles)
```
