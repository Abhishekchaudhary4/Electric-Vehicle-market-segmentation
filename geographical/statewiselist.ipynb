```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('geoMap.csv')
```

```python
data.head()
```

```python
data['electric vehicle: (09/10/2022 - 09/10/2023)'].mean()
```

```python
data['electric vehicle: (09/10/2022 - 09/10/2023)'].max()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(data['Region'], data['electric vehicle: (09/10/2022 - 09/10/2023)'])
plt.xlabel('Region')
plt.ylabel('Electric Vehicle Sales')
plt.title('Electric Vehicle Sales by Region')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability

plt.tight_layout()
plt.show()
```
