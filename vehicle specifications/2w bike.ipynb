```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df1['Owned for'] = label_encoder.fit_transform(df1['Owned for'])
```

```python
df1.head()
```

```python
df1['Ridden for'] = label_encoder.fit_transform(df1['Ridden for'])
```

```python
df1 = pd.get_dummies(df1, columns=['Model Name'])
```

```python
df1.head()
```

```python
df1['Visual Appeal'].fillna(df1['Visual Appeal'].mean(), inplace=True)
df1['Reliability'].fillna(df1['Reliability'].mean(), inplace=True)
df1['Performance'].fillna(df1['Performance'].mean(), inplace=True)
df1['Comfort'].fillna(df1['Comfort'].mean(), inplace=True)
df1['Maintenance cost'].fillna(df1['Maintenance cost'].mean(), inplace=True)
df1['Value for Money'].fillna(df1['Value for Money'].mean(), inplace=True)

# Continue this for other columns as well.
```

```python
df1['Service Experience'].fillna(0, inplace=True)
df1['Extra Features'].fillna(0, inplace=True)
# Continue this for other columns as well.
```

```python
most_common_ridden_for = df1['Ridden for'].mode().iloc[0]
df1['Ridden for'].fillna(most_common_ridden_for, inplace=True)
```

```python
import matplotlib.pyplot as plt

# Group the data by "Model Name" and calculate the maximum ownership duration for each model.
max_owned_for = df.groupby('Model Name')['Owned for'].max()

# Find the model with the maximum ownership duration.
max_ownership_duration = max_owned_for.max()
model_with_max_ownership = max_owned_for[max_owned_for == max_ownership_duration].index[0]

# Create a bar chart to visualize the maximum ownership duration for each model.
plt.figure(figsize=(10, 6))
plt.bar(max_owned_for.index, max_owned_for)
plt.xlabel("Vehicle Model")
plt.ylabel("Maximum Ownership Duration")
plt.title(f"Vehicle with Maximum Ownership Duration: {model_with_max_ownership} ({max_ownership_duration})")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Maintenance cost'], df['rating'])
plt.xlabel('Maintenance Cost')
plt.ylabel('Rating')
plt.title('Maintenance Cost vs. Rating')
plt.show()
```

```python
usage_counts = df['Used it for'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(usage_counts, labels=usage_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Usage Purpose Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df['Maintenance cost'].dropna())
plt.ylabel('Maintenance Cost')
plt.title('Distribution of Maintenance Costs')
plt.show()
```

```python
attributes = ['Visual Appeal', 'Reliability', 'Performance', 'Service Experience']
for attribute in attributes:
    plt.figure(figsize=(12, 12))
    plt.bar(df['Model Name'], df[attribute])
    plt.xlabel('Vehicle Model')
    plt.ylabel(attribute)
    plt.title(f'Average {attribute} by Model')
    plt.xticks(rotation=90)
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['rating'], bins=5, edgecolor='k')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()
```
