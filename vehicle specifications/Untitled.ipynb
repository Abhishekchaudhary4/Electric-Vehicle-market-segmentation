```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('ElectricCarData_Clean.csv')
```

```python
data
```

```python
selected_columns = data.iloc[:, 2:8]  # Select columns 2 to 7 (0-based index)

# Calculate the mean (average) for each column
averages = selected_columns.mean()

# Print the averages for each column
print(averages)
```

```python
# Creating a histogram for the 'Range_Km' column
plt.figure(figsize=(8, 6))
plt.hist(data['Range_Km'], bins=20, color='skyblue')
plt.title('Range Distribution (Kilometers)')
plt.xlabel('Range (Kilometers)')
plt.ylabel('Frequency')
plt.show()
```

```python
import seaborn as sns
# Example: Creating a scatter plot for 'PriceEuro' vs. 'Range_Km'
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='PriceEuro', y='Range_Km', hue='Segment')
plt.title('Price vs. Range')
plt.xlabel('Price (Euros)')
plt.ylabel('Range (Kilometers)')
plt.legend(title='Segment')
plt.show()
```

```python
# Example: Creating a bar plot for the 'Brand' column
plt.figure(figsize=(12, 8))
brand_counts = data['Brand'].value_counts()
brand_counts.plot(kind='bar', color='salmon')
plt.title('Number of Models by Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Models')
plt.xticks(rotation=90)
plt.show()
```

```python
# Example: Creating a pair plot for selected columns
columns_to_plot = ['Range_Km', 'Efficiency_WhKm', 'PriceEuro']
sns.set(style="ticks")
sns.pairplot(data=data, vars=columns_to_plot, hue='Segment', palette='viridis')
plt.suptitle('Pair Plot of Range, Efficiency, and Price')
plt.show()
```

```python
# Example: Creating a box plot for the 'Range_Km' column
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Segment', y='Range_Km', palette='pastel')
plt.title('Range Distribution by Segment')
plt.xlabel('Segment')
plt.ylabel('Range (Kilometers)')
plt.xticks(rotation=45)
plt.show()
```

```python
columns_to_plot = ['TopSpeed_KmH', 'Range_Km', 'Efficiency_WhKm', 'FastCharge_KmH', 'PriceEuro']

# Create a grouped bar plot
plt.figure(figsize=(12, 8))

for col in columns_to_plot[:-1]:
    plt.bar(data['Model'], data[col], label=col)

plt.xlabel('Vehicle Model')
plt.ylabel('Values')
plt.title('Vehicle Data vs. Price')
plt.xticks(rotation=90)
plt.legend(loc='upper left')
plt.show()
```

```python
columns_to_plot = ['TopSpeed_KmH', 'Range_Km', 'Efficiency_WhKm', 'FastCharge_KmH']

# Create separate bar plots for each column
for col in columns_to_plot:
    plt.figure(figsize=(14, 10))
    plt.bar(data['Model'], data[col], color='skyblue')
    plt.xlabel('Vehicle Model')
    plt.ylabel(col)
    plt.title(f'{col} vs. PriceEuro')
    plt.xticks(rotation=90)
    plt.show()
```

```python
columns_to_plot = ['TopSpeed_KmH', 'Range_Km', 'Efficiency_WhKm', 'FastCharge_KmH']

# Create line plots for each column
plt.figure(figsize=(12, 8))

for col in columns_to_plot:
    sns.lineplot(data=data, x='PriceEuro', y=col, label=col)

plt.xlabel('PriceEuro')
plt.ylabel('Values')
plt.title('PriceEuro vs. Vehicle Data')
plt.legend(loc='upper right')
plt.show()
```
